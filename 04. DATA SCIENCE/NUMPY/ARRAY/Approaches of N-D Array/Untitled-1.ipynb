```python
import numpy as np
import sys
lst = [10,20,30,40,50,60,70,80,90]
print(lst,type(lst))
lstSize = sys.getsizeof(lst)
print(lstSize)
a = np.array(lst)
print(a,type(a))
aSize = sys.getsizeof(a)
print(aSize,)
```
