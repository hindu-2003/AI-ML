```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset

data = pd.read_csv('advertising.csv')
```

```python
# Display the first few rows of the dataset

print(data)
```

```python
# Check for missing values

print(data.isnull().sum())
```

```python

# Visualize the relationship between features and target variable
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.scatter(data['TV'], data['sales'])
plt.title('TV vs Sales')

plt.subplot(1, 3, 2)
plt.scatter(data['radio'], data['sales'])
plt.title('Radio vs Sales')

plt.subplot(1, 3, 3)
plt.scatter(data['newspaper'], data['sales'])
plt.title('Newspaper vs Sales')

plt.show()
```

<b>Relationship Between Advertising Channels and Sales</b>


<b>TV Advertising</b>:

The scatter plot of <b>TV vs Sales</b> shows a strong positive linear relationship. 

This suggests that increasing TV advertising spending is likely to lead to higher sales.

The coefficient for TV in the linear regression model is positive and significant, indicating that TV advertising has a strong impact on sales.


<b>Radio Advertising</b>:

The scatter plot of <b>Radio vs Sales</b> also shows a positive relationship, though it may not be as strong as TV. 

This suggests that radio advertising contributes to sales but to a lesser extent compared to TV.

The coefficient for <b>Radio</b> is positive but smaller than that of TV, indicating a moderate impact on sales.



<b>Newspaper Advertising</b>:

The scatter plot of <b>Newspaper vs Sales</b> shows a weaker or less clear relationship. 

This suggests that newspaper advertising may have a minimal or negligible impact on sales.

The coefficient for Newspaper is the smallest among the three, indicating that it has the least impact on sales.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

for i, col in enumerate(['TV', 'radio', 'newspaper'], 1):
    plt.subplot(1, 3, i)
    plt.boxplot(data[col])
    plt.title(col)

plt.show()
```

---

```python
# Define features (X) and target variable (y)

X = data[['TV', 'radio', 'newspaper']]

y = data['sales']
```

```python
X
```

```python
y
```

```python

from sklearn.preprocessing import StandardScaler

# Normalize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
y
```

```python

from sklearn.model_selection import train_test_split
# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Train the model

model.fit(X_train, y_train)
```

```python
# Make predictions on the test set

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error (MSE) and R-squared (R2) score
mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
```

### Model Performance

#### Mean Squared Error (MSE):

The MSE measures the average squared difference between the actual and predicted sales.

A lower MSE indicates better model performance.

For example, if the MSE is 2.8, it means the model's predictions are, on average, off by about 2.8 units squared.

#### R-squared (R):

The R score measures the proportion of variance in the target variable (sales) that is explained by the model.

It ranges from 0 to 1, where 1 indicates a perfect fit.

If the R score is 0.89, it means that 89% of the variability in sales is explained by the advertising channels (TV, Radio, Newspaper).

---

```python
# Visualize the predictions vs actual values

plt.scatter(y_test, y_pred)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.show()

```

```python
# Print the coefficients
print('Coefficients:', model.coef_)

print('Intercept:', model.intercept_)
```

### Key Insights


<b>TV Advertising is the Most Influential</b>:

The coefficient for TV advertising is the largest (3.83), indicating that TV has the strongest positive impact on sales. Increasing TV advertising spending will yield the highest return in terms of sales.

<b><b>Radio Advertising is Also Effective</b>:

The coefficient for Radio advertising is 2.80, which is significant but smaller than TV. 

This suggests that Radio advertising is effective but less impactful than TV.

<b>Newspaper Advertising is Least Effective</b>:

The coefficient for Newspaper advertising is very small (0.06), indicating that it has almost no impact on sales.

This suggests that Newspaper advertising may not be a worthwhile investment.

<b>Baseline Sales</b>:

The intercept (14.04) represents the baseline sales when no advertising is done (or when advertising is at its mean value). 

This is the expected sales value in the absence of any advertising efforts.

### Key Takeaways


#### TV Advertising is the Most Effective: 

TV advertising has the strongest positive impact on sales, as indicated by its high coefficient and strong linear relationship in the scatter plot.

#### Radio Advertising is Moderately Effective: 

Radio advertising contributes to sales but is less impactful than TV.

#### Newspaper Advertising is Least Effective: 

Newspaper advertising has a minimal or negligible impact on sales, as indicated by its small coefficient and weak relationship in the scatter plot.

#### Model Performance is Good: 

The high R score (e.g., 0.89) indicates that the model explains a large portion of the variability in sales, and the low MSE suggests accurate predictions.
