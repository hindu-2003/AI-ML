```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

data = pd.read_csv("advertising.csv")

print(data.head())
```

```python

print("\nMissing values:\n", data.isnull().sum())
```

```python
#  Visualize Relationships

plt.figure(figsize=(12, 6))

plt.subplot(1,3,1)
plt.scatter(data["TV"], data["sales"])
plt.title("TV vs Sales")

plt.subplot(1,3,2)
plt.scatter(data["radio"], data["sales"])
plt.title("Radio vs Sales")

plt.subplot(1,3,3)
plt.scatter(data["newspaper"], data["sales"])
plt.title("Newspaper vs Sales")

plt.show()
```

```python

#  Boxplots (Outlier Detection)

plt.figure(figsize=(12, 6))
for i, col in enumerate(['TV', 'radio', 'newspaper'], 1):
    plt.subplot(1, 3, i)
    plt.boxplot(data[col])
    plt.title(col)
plt.show()
```

```python

#  Outlier Removal using IQR

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Remove outliers for all 3 columns
clean_data = data.copy()
for col in ['TV', 'radio', 'newspaper']:
    clean_data = remove_outliers_iqr(clean_data, col)

print("\nOriginal Shape:", data.shape)
print("After Outlier Removal:", clean_data.shape)
```

```python
#  Feature Matrix & Target

X = clean_data[['TV', 'radio', 'newspaper']]
y = clean_data['sales']
```

```python
# Normalize

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#  Train Model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict

y_pred = model.predict(X_test)
```

```python
#  Evaluation

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"\nRMSE After Removing Outliers: {rmse}")
print(f"R Score After Removing Outliers: {r2}")
```

```python

#  Visualization: Actual vs Predicted

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted (After Outlier Removal)")
plt.show()

# Coefficients
print("\nCoefficients:", model.coef_)
print("Intercept:", model.intercept_)
```


###  Test With New Unknown Data 


```python

new_data = pd.DataFrame({
    'TV': [350, 50],
    'radio': [90, 10],
    'newspaper': [2, 2]
})

print(new_data)

# Scale features
new_scaled = scaler.transform(new_data)

# Predict
new_predictions = model.predict(new_scaled)

print("\nPredicted Sales for New Entries:")
for i, pred in enumerate(new_predictions):
    print(f"Entry {i+1}: Predicted Sales = {pred:.2f}")
```
