```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

for i, feature in enumerate(['rm', 'lstat', 'medv']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=boston[feature])
    plt.title(f"Boxplot of {feature}")

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.20)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Apply to key features
for col in ['rm', 'lstat', 'medv']:
    boston = remove_outliers(boston, col)
```

## R square and adjusted R square


Formula

**R^2 = 1 - SSR/SST**


R^2	=	coefficient of determination
SSR	=	sum of squares of residuals
SST	=	total sum of squares


```python
from sklearn.metrics import r2_score

score=r2_score(y_test,y_pred)
print(score)
```

**Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]**

where:

R2: The R2 of the model
n: The number of observations
k: The number of predictor variables


```python
#display adjusted R-squared

n = X_test.shape[0]
k = X_test.shape[1]

adjusted_r2 = 1 - (1 - score) * (n - 1) / (n - k - 1)

print("Adjusted R:", round(adjusted_r2, 4))
```

### Multicollinearity

Multicollinearity occurs when two or more independent features are highly correlated with each other.

Variance Inflation Factor (VIF) quantifies how much the variance of a regression coefficient is inflated due to multicollinearity among the independent variables.

    In short: VIF tells you how badly a feature is correlated with the others.

### Check Multicollinearity with VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X_vif = add_constant(pd.DataFrame(X))  # add intercept
vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X_vif.values, i+1) for i in range(len(X.columns))]
print(vif)
```

VIF Value	Interpretation

    1 - 5	Acceptable
    
    > 5	Moderate multicollinearity
    
    > 10	Severe multicollinearity  consider removing or combining variables

### Comparitive study

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

models = {
    "Linear": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.01),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "SVR": SVR()
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results.append((name, round(r2, 4), round(rmse, 4)))


results_df = pd.DataFrame(results, columns=["Model", "R Score", "RMSE"])
print("\n Model Comparison:")
print(results_df)
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.barplot(x="R Score", y="Model", data=results_df.sort_values("R Score", ascending=False))
plt.title("Model R Score Comparison")
plt.xlim(0, 1)
plt.show()
```

### Comparitive study

```python
results_df_melted = results_df.melt(id_vars="Model", var_name="Metric", value_name="Score")

plt.figure(figsize=(10, 6))
sns.barplot(x="Score", y="Model", hue="Metric", data=results_df_melted, palette="muted")
plt.title("Model Comparison: R Score vs RMSE")
plt.xlim(0, results_df_melted['Score'].max() + 1)
plt.legend(loc='best')
plt.show()
```
