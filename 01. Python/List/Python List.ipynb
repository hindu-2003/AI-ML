| Feature          | `list.sort()`                                   | `sorted()`                                                       |
| ---------------- | ----------------------------------------------- | ---------------------------------------------------------------- |
| **Type**         | Method of a list object                         | Built-in function                                                |
| **Return Value** | **In-place sorting** → returns `None`           | **Creates a new list** → returns sorted iterable                 |
| **Mutability**   | Modifies the **original list**                  | Leaves the original iterable **unchanged**                       |
| **Usable On**    | Only **lists**                                  | Works on **any iterable** (list, tuple, dict, set, string, etc.) |
| **Syntax**       | `list.sort(key=None, reverse=False)`            | `sorted(iterable, key=None, reverse=False)`                      |
| **Performance**  | Faster (no extra list creation, sorts in-place) | Slightly slower (creates a new list before sorting)              |
| **Flexibility**  | Limited to lists                                | Works on iterables beyond lists                                  |
| **Chaining**     | Cannot be chained (since it returns `None`)     | Can be chained (returns a new sorted list)                       |

### Need to check

| Feature          | `list.reverse()`                                 | `reversed()`                                                               |
| ---------------- | ------------------------------------------------ | -------------------------------------------------------------------------- |
| **Type**         | Method of a list object                          | Built-in function                                                          |
| **Return Value** | **In-place reversal** → returns `None`           | Returns an **iterator** (not a list)                                       |
| **Mutability**   | Modifies the **original list**                   | Does **not modify** the original object                                    |
| **Usable On**    | Only **lists**                                   | Works on **any sequence** (list, tuple, string, range, etc.)               |
| **Chaining**     | Cannot be chained (returns `None`)               | Can be used in expressions, loops, or wrapped with `list()` to materialize |
| **Performance**  | Faster (direct in-place reversal, no new object) | Slightly slower (creates an iterator)                                      |
| **Syntax**       | `list.reverse()`                                 | `reversed(sequence)`                                                       |

### Programs on Lists

#### Multiplies all the items in a list

```python
def multiply_list(items):
    tot = 1
    for x in items:
        tot = tot * x
    return tot
    
res = multiply_list([1,2,5,10])
print(res)
```

#### Get the largest number from a list

```python
def max_num_in_list( list ):
    
    max = list[0]
    
    for a in list:
        if a > max:
            max = a
    return max
    
print(max_num_in_list([2,17,1,3,4,15]))
```

#### Function to list of words that are longer than n from a given list of words.

```python
def long_words(n, str):
    
    word_len = []
    
    txt = str.split(" ")
    
    for x in txt:
        if len(x) > n:
            word_len.append(x)
            
    return word_len
    
print(long_words(4, "The quick brown fox jumps over the lazy dog"))
```

#### fun to count no of occurances of an element 

```python
def ele_count(l1, x): 
    count = 0
    for ele in l1: 
        if (ele == x): 
            count = count + 1
    return count 

l= [1,2,3,4,4,4,5]
x = 4

print('{} appeared for {} times'.format(x, ele_count(l, x))) 
```

```python
# Function that takes two lists and returns True if they have at least one common

def common_data(list1, list2):
    result = False
    for x in list1:
         for y in list2:
            if x == y:
                result = True
    return result
print(common_data([1,2,3,4,5], [5,6,7,8,9]))
```
