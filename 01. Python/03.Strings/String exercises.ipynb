##### Function to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself.

```python
def change_char(str1):
    char = str1[0]
    str1 = str1.replace(char, '$')
    #print(str1)
    str1 = char + str1[1:]

    return str1

print(change_char('Manasa Madam'))
```

###### Function to get a single string from two given strings, separated by a space and swap the first two characters of each string.

```python
def chars_mix_up(a, b):
    new_a = b[:2] + a[2:]
    new_b = a[:2] + b[2:]

    return new_a + ' ' + new_b

print(chars_mix_up('Python', 'Program'))
```

##### Python function that takes a list of words and return the longest word and the length of the longest one.

```python
def find_longest_word(words_list):
    word_len = []
    
    for n in words_list:
        word_len.append((len(n), n))
    print(word_len)
    
    word_len.sort()
    print(word_len)

    return word_len[-1][0], word_len[-1][1]

result = find_longest_word(["String", "List", "Dictionary"])
print("Longest word: ",result[1])
print("Length of the longest word: ",result[0])
```

```python
def find_longest_word(words_list):
   
    longest_word = max(words_list, key=len)
    return len(longest_word), longest_word

result = find_longest_word(["String", "List", "Dictionary"])
print("Longest word:", result[1])
print("Length of the longest word:", result[0])
```

##### Python program to remove characters that have odd index values in a given string.

```python
def odd_values_string(str):
    result = "" 
    for i in range(len(str)):
        if i % 2 == 0:
            result = result + str[i]
    return result

print(odd_values_string('python'))
```

##### Python function to get a string made of 4 copies of the last two characters of a specified string

```python
def insert_end(str):
	sub_str = str[-2:]
	return sub_str * 4

print(insert_end('Python'))
```

###### Python function to get a string made of the first three characters of a specified string.
If the length of the string is less than 3, return the original string.

```python
def first_three(str):
    if len(str) > 3:
        return str[:3] 
    else:
        return str
    
print(first_three('Python'))
print(first_three('Py'))
```

###### Sort a string lexicographically

```python
def alpha_sort(s):
    return sorted(s)

print(alpha_sort('python'))
```

##### Reverse a string

```python
def reverse_string(str1):
    return ''.join(reversed(str1))

print(reverse_string("abcdef"))
```

```python
def reverse_string(str1):
    return list(reversed(str1))

print(reverse_string("abcdef"))
```

```python
def reverse_string(str1):
    return str1[::-1]

print(reverse_string("abcdef"))
```

##### Reverse words in a string

```python
def reverse_string_words(text):
    words = text.split(" ")
   
    return words[::-1]
    
print(reverse_string_words("The quick brown fox jumps over the lazy dog"))
```

###### Remove spaces from a given string

```python
def remove_spaces(str1):
    str1 = str1.replace(' ','')
    return str1
    
print(remove_spaces("p y t h o n p r g"))
```

##### Compute sum of digits of a given string

```python
def sum_digits_string(str1):
    sum_digit = 0
    for x in str1:
        if x.isdigit() == True:
            z = int(x)
            sum_digit = sum_digit + z

    return sum_digit
     
print(sum_digits_string("12xyz3abcd45"))
```
