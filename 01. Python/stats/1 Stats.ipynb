# Central tendency

Central tendency is a statistical measure that identifies the center or typical value of a dataset. 

In simple words, it tells us where most of the data points are clustered.

It’s a way to summarize data with a single representative value.

```python
import numpy as np 
import numpy.random as rnd
```

```python
x= np.reshape(np.arange(20.0),(4,5))
x
```

```python
np.mean(x)
```

```python
x= rnd.randn(4,5)
x
```

```python
np.median(x)
```

```python
x = [1,2,3,4,500]

print(np.mean(x))
print(np.median(x))

# if the data very good and cleaned data :  Mean

# when we have outlier or noise prefer: median
```

```python
import statistics

x = [34, 43, 81, 106, 106,115]
 
print("Mode of given data set is % s" % (statistics.mode(x)))
```

```python
import numpy as np
from scipy import stats

data = np.array([34, 43, 81, 106, 106,115])
mode_result = stats.mode(data)

mode_result
```

```python
print("Mode:", mode_result.mode)  
print("Count:", mode_result.count) 
```

```python
import numpy as np

data = np.array([34, 43, 81, 106, 106,115])

values, counts = np.unique(data, return_counts=True)
print(values,counts)
```

```python
np.argmax(counts)
```

```python
mode = values[np.argmax(counts)]

print("Mode:", mode) 
```

# Measures of dispersion

Describe the spread or variability of data around the central tendency.

While central tendency tells us the “center” of data, dispersion tells us how spread out the data is.

```python
x= [44,50,38,96,42,47,40,39,46,50]

print(np.mean(x))
print(np.std(x))
print(np.var(x))
```

```python
max = np.mean(x)+np.std(x)
min = np.mean(x)-np.std(x)

print(round(max,2),round(min,2))
```

##### find out outlier in above example

```python
from statistics import variance, stdev
import numpy as np

game_points = np.array([35,56,43,59,63,79,35,41,64,43,93,60,77,24,82])
game_points
```

```python
print(' Calculate Range')

dt_rng = np.max(game_points,axis=0) - np.min(game_points,axis=0) 
print ("Range:",dt_rng)
```

```python

dt_qntls = np.percentile(game_points,25)

dt_qntls
```

```python
l = [25,50,75]

for val in l :
    
    dt_qntls = np.percentile(game_points,val)
    
    print (str(val)+"%" ,dt_qntls)
```

```python
print (" Calculate IQR:")

q75, q25 = np.percentile(game_points, [75 ,25])

print ("Interquartile range:",q75-q25)
```


    Mean ± SD → standard for normally distributed data.

    Median ± IQR → better for skewed or outlier-prone data.

    Always check both: central tendency tells “typical value,” dispersion tells “reliability/variability.”

# Central Limit Theorem

```python
import numpy as np
from matplotlib import pyplot

rolls = np.random.randint(1, 7, 50)
print(rolls)

print(np.mean(rolls))
```

```python
means = [np.mean(np.random.randint(1, 7, 50)) for i in range(100000)]

pyplot.hist(means)
pyplot.show()
```
