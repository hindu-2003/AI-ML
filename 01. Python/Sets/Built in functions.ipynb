### map()

```python
nums = [1, 2, 3, 4, 5]  

def sq(n):    
    return n*n  

square3 = list(map(sq, nums))
print(square3)
```

```python
nums = [1, 2, 3, 4, 5]  
def sq(n):    
    return n*n  

square1 = map(sq, nums)

print(tuple(square1))
```

```python
# Square each item of the list
def square(x):
    return x*2

lst = [1, 2, 3, 4, 5]

newlst = list(map(square, lst))
print(newlst)
```

### filter()

```python
alphabets = ['a', 'b', 'd', 'a', 'e', 'i', 'j', 'o']

def filterVowels(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']

    if(alphabet in vowels):
        return True
    else:
        return False

filteredVowels = set(filter(filterVowels, alphabets))
print(filteredVowels)
```

```python
def checkAge(age):
    if age >= 18:
        return True
    else:
        return False

lst = [10,14,18,22,24]

adults = filter(checkAge, lst)
print(list(adults))
```

### enumerate

### sorted()

```python
a = [1,4,3,2,8]
print(sorted(a))

print(sorted(a, reverse=True))
```

```python
c = ['a','z','A','b']

print(sorted(c, reverse=True))
```

```python
words = ['banana', 'abc', 'dictionary']

print(sorted(words, key=len))     
```

```python
pairs = [('Alice', 85), ('Bob', 90), ('Eve', 80)]

print(sorted(pairs, key=lambda x: x[1]))  
```

```python
names = ['Guido van Rossum', 'Bjarne Stroustrup' , 'James Gosling']

print(sorted(names, key= lambda name: name.split()[-1]))
```

### Max, Min

```python
studmarks = [('ABC', 35), ('CDE', 25), ('XYZ', 30),('PQR', 20)]

maxlst = max(studmarks, key=lambda student: student[1])

minlst = min(studmarks, key=lambda student: student[1])

print(maxlst)
print(minlst)
```
