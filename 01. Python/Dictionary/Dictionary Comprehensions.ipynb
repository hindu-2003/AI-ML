### Dictionary Comprehension

```python
square_dict = {num:num*num for num in range(1, 6)}
print(square_dict)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

double_dict1 = {k:v*2 for (k,v) in dict1.items()}

print(double_dict1)
```

#### list elements as the keys and the length of each string as the values

```python
dict_comp = ["allow","sequences","to","be","built","from","other","sequences"]

d = {f:len(f) for f in dict_comp}
print(d)
```

```python
# Convert the first char of a List of strings, first letter capitalized 

d1 = {f:f.capitalize() for f in dict_comp}
print(d1)
```

```python
{f:f.upper() for f in dict_comp}
```

```python
scores = {'maths': 55, 'physics': 65, 'chemistry': 75}
agg = 1.25
new_marks = {sub: marks * agg for (sub, marks) in scores.items()}
print(new_marks)
```

```python
scores = {'maths': 50, 'physics': 60, 'chemistry': 75}
new_marks = {sub: marks  for (sub, marks) in scores.items() if marks%2==0}
print(new_marks)
```

```python
scores = {'maths': 50, 'physics': 60, 'chemistry': 75}
new_marks = {sub: marks  for (sub, marks) in scores.items() if marks%2==0 if marks >50}
print(new_marks)
```
