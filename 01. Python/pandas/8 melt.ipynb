### MELT

The <b>melt() </b> function in Pandas is like a magic wand for your data. 

It allows you to transform data from a wide format into a long format, or in simpler terms.

        it helps you make your data more organized and ready for analysis.

Imagine you have a table with lots of columns, and each column represents something different. 

This wide table format as we have seen above is okay for human eyes, but it's not so great when you want to do some serious data analysis or create cool visualizations.

```python
import pandas as pd
df = pd.read_csv("weather.csv")
df
```

Why is this useful?

    Easier to plot grouped charts (city vs. temperature).
    
    Easier to groupby & aggregate (e.g., average per city).

id_vars = keep these columns fixed.

var_name = new column name for melted column headers.

value_name = new column name for melted cell values.

```python
melted = pd.melt(df, id_vars=["day"], var_name='city', value_name='temperature')
melted
```

pd.melt() reshapes data from wide format → long (tidy) format.

    Wide format = one column per variable (your original DataFrame: chicago, chennai, berlin).
    
    Long format = one row per observation (city–temperature pairs).

### Line plot by city

```python
import matplotlib.pyplot as plt

for city, group in melted.groupby("city"):
    plt.plot(group["day"], group["temperature"], marker="o", label=city)

plt.title("Weekly Temperatures")
plt.ylabel("Temperature (°F)")
plt.legend()
plt.show()
```
