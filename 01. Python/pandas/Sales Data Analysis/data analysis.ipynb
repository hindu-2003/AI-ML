### More visualizations

```python
plt.figure(figsize=(10,6))

data1.groupby('Gender')['Sales in Rs Laks'].mean().plot(kind='bar')

plt.title('Average Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Sales (Rs Laks)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
data1.groupby(['Location', 'Gender'])['Sales in Rs Laks'].mean().unstack().plot(kind='bar')
plt.title('Average Sales by Location and Gender')
plt.ylabel('Average Sales (Rs Laks)')
plt.show()
```

### Sales by Age, colored by Gender

```python
import matplotlib.pyplot as plt

colors = {'Male': 'blue', 'Female': 'red'}
plt.figure(figsize=(8,6))
plt.scatter(data1['Age'], data1['Sales in Rs Laks'], 
            c=data1['Gender'].map(colors))
plt.xlabel('Age')
plt.ylabel('Sales (Rs Laks)')
plt.title('Sales vs Age by Gender')
plt.legend(colors)
plt.grid(True)
plt.show()
```
