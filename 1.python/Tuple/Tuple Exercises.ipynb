### Tuple exercises

#### function that calculates the sum and product of all elements in a tuple of numbers.

```python
def sum_product(t):
    
    sum_result = 0
    product_result = 1
 
    for num in t:
        sum_result += num
        product_result *= num
 
    return sum_result, product_result
 
input_tuple = (1, 2, 3, 4)
```

```python
sum_result, product_result = sum_product(input_tuple)
print(sum_result, product_result)  
```

#### function that takes two tuples and returns a tuple containing the element-wise sum of the input tuples

```python
def tuple_elementwise_sum(t1, t2):
    result = tuple(a + b for a, b in zip(t1, t2))
    return result
```

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

output_tuple = tuple_elementwise_sum(tuple1, tuple2)
print(output_tuple)  
```

#### function that takes a tuple and a value, and returns a new tuple with the value inserted at the beginning of the original tuple and end also.

```python
def insert_value_at_beginning(input_tuple, value_to_insert):
    
    return (value_to_insert,) + input_tuple + (value_to_insert,) 
```

```python
input_tuple = (2, 3, 4)
value_to_insert = 1
```

```python
output_tuple = insert_value_at_beginning(input_tuple, value_to_insert)                        
print(output_tuple)  
```

#### function that takes a tuple of strings and concatenates them, separating each string with a space.

```python
input_tuple = ('Hello', 'World', 'from', 'Python')

def concatenate_strings(input_tuple):
    
    return ' '.join(input_tuple)
```

```python
output_string = concatenate_strings(input_tuple)

print(output_string)  
```

### function that takes a tuple of tuples and returns a tuple containing the diagonal elements of the input.

```python
input_tuple = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

def get_diagonal(input_tuple):
    
    return tuple(input_tuple[i][i] for i in range(len(input_tuple)))
```

```python
output_tuple = get_diagonal(input_tuple)

print(output_tuple)  
```

#### function that takes two tuples and returns a tuple containing the common elements of the input tuples.

```python
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (4, 5, 6, 7, 8)

def common_elements(tuple1, tuple2):
    
    return tuple(set(tuple1) & set(tuple2))
```

```python
output_tuple = common_elements(tuple1, tuple2)

print(output_tuple)  
```

### Remove an empty tuple(s) from a list of tuples

```python
L = [(), (),('a', 'b'), ('a', 'b', 'c'), ('d')]

L = [t for t in L if t]
print(L) 
```

### Python program to get the 4th element from the last element of a tuple.

```python
# Get the 4th element of the tuple (index 3)
item = tuplex[3]
print(item)
```

```python
# Get the 4th element from the end of the tuple (index -4)
item1 = tuplex[-4]
print(item1) 
```

###  Tuple to a string

```python
tup = ('p', 'y', 't', 'h', 'o', 'n', ' ', 't', 'u', 'p', 'l', 'e')

# Use the 'join' method to concatenate the characters in the tuple without any spaces and create a single string
str = ''.join(tup)
print(str)
```

#### Python program to find repeated items in a tuple.

```python
tuplex = 2, 4, 5, 6, 2, 3, 4, 4, 7
print(tuplex)

# Count the number of times the value 4 appears in the 'tuplex' tuple
count = tuplex.count(4)
print(count) 
```
