<h1 style="color:blue">Reshape dataframe using stack/unstack</h1>

The stack() function is used to reshape a DataFrame by converting (or "pivoting") columns into rows. 

This operation is particularly useful when you have a DataFrame with multi-level column headers and you want to "stack" the data into a more compact, long-format DataFrame.

stack() converts columns into rows, resulting in a more compact, long-format DataFrame.

unstack() does the opposite, converting rows back into columns.

These functions are particularly useful when working with multi-level indices.

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
df
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.stack()
```

```python
df.stack(level=0)
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df_stacked.unstack()
```
