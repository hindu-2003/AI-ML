```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("titanic_train.csv")
df
```

### Group by Sex & Pclass → Mean Age + Total Survivors

```python

table = pd.pivot_table(
    df,
    index=['Sex', 'Pclass'],
    aggfunc={'Age': np.mean, 'Survived': np.sum}
)
```

```python
table
```

```python

table1 = df.groupby(['Sex', 'Pclass']).agg({'Age': 'mean', 'Survived': 'sum'})
```

```python
table1
```

### Mean Survival Rate by Sex & Pclass

```python

table2 = pd.pivot_table(
    df,
    index=['Sex', 'Pclass'],
    values='Survived',
    aggfunc=np.mean
)

table2
```

```python
table3 = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
```

```python
table3
```

### Survivors by Sex vs Pclass (Matrix-style)

```python
table4 = pd.pivot_table(
    df,
    index='Sex',
    columns='Pclass',
    values='Survived',
    aggfunc=np.sum,
    fill_value=0
)

table4
```

```python
table5 = df.groupby(['Sex', 'Pclass'])['Survived'].sum().unstack(fill_value=0)
table5
```

### Handling Missing Data (Age mean by Sex, Survived, Pclass × Embarked)

```python
table6 = pd.pivot_table(
    df,
    index=['Sex', 'Survived', 'Pclass'],
    columns='Embarked',
    values='Age',
    aggfunc=np.mean,
    fill_value=np.round(df['Age'].mean(), 1)
)
table6
```

```python
table7 = (
    df.groupby(['Sex', 'Survived', 'Pclass', 'Embarked'])['Age']
      .mean()
      .unstack(fill_value=np.round(df['Age'].mean(), 1))
)
table7
```
