### Binning

Binning (bucketing) groups numeric values into intervals (bins) and replaces each value with a representative for that bin.

Purpose:

    Reduce the effect of small observation errors/noise (data smoothing)
    
    Convert continuous variables into categorical features
    
    Improve performance of some models (e.g., decision trees, logistic regression)
    
    Improve interpretability

```python
import pandas as pd
df = pd.read_csv('cupcake.csv')
df
```

```python
min_value = df['Cupcake'].min()
max_value = df['Cupcake'].max()
print(min_value)
print(max_value)
```

```python
labels = ['small', 'medium']
```

We can use the cut() function to convert the numeric values of the column Cupcake into the categorical values. 

We need to specify the bins and the labels. 

In addition, we set the parameter include_lowest to True in order to include also the minimum value.

```python
df['bins'] = pd.cut(df['Cupcake'], bins=2, labels=labels, include_lowest=True)
```
