###  Traverse through a dictionary

```python
def traverseDict(d):
    
    for key, value in d.items():
        print(key, value)

myDict = {'name': 'Edy', 'age': 26}
```

```python
traverseDict(myDict)
```

###  Searching for an element in a dictionary

```python
myDict = {'name': 'Edy', 'age': 27}

def searchDict(d, value):
    for key, val in d.items():
        if val == value:
            return key, val
    return 'The value does not exist'
```

```python
print(searchDict(myDict, 28))
```

```python
print(searchDict(myDict, 'Edy'))
```

### sorted method

```python
myDict = {'eooooa': 2, 'aas': 3, 'udd': 1, 'sseo': 4, 'werwi': 5}
print(sorted(myDict, key=len))
```

### Sort dictionary keys by length and get a new dictionary

```python
sorted_keys= sorted(myDict, key=len)
sorted_keys
```

```python
sorted_dict = {k: myDict[k] for k in sorted_keys}
print(sorted_dict)
```

### Differences and similarities sort() vs sorted()

The primary difference between the two is that list.sort() will sort the list in-place, mutating its indexes and returning None,

whereas sorted() will return a new sorted list leaving the original list unchanged. 

Another difference is that sorted() accepts any iterable while list.sort() is a method of the list class and can only be used with lists.

### Define a function which takes a dictionary as a parameter and returns the key with the highest value in a dictionary.

```python
# max() function
my_dict = {'a': 5, 'b': 9, 'c': 2}

def max_value_key(my_dict):
    return max(my_dict, key=my_dict.get)
```

```python
max_value_key(my_dict)
```

The max() function returns the largest item in an iterable. 

It can also be used to find the largest item between two or more parameters.

##### key parameter so that we can find the dictionary's key having the largest value.

# Dict comprehension

```python
# Define a function which takes as a parameter dictionary and returns a dictionary in which everse the key-value pairs are reversed.

my_dict = {'a': 1, 'b': 2, 'c': 3}

def reverse_dict(my_dict):
    return {v: k for k, v in my_dict.items()}
```

```python
reverse_dict(my_dict)
```

```python
# Define a function that takes a dictionary as a parameter and returns a dictionary with elements based on a condition.

def filter_dict(my_dict, condition):
    return {k: v for k, v in my_dict.items() if condition(k, v)}
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4} 

filtered_dict = filter_dict(my_dict, lambda k, v: v % 2 == 0) 

print(filtered_dict)
```
