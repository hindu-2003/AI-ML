```python
# The isdisjoint() method returns True if two sets don't have any common items between them, 
# i.e. they are disjoint.  Else the returns False.

s1 = {1, 2, 3}
s2 = {4, 5, 6}

# checks if set s1 and set s2 are disjoint
print(s1.isdisjoint(s2))
```

```python
s1 = {1, 2, 3}
s2 = {4, 5, 3, 6}

# checks if set s1 and set s2 are disjoint
print(s1.isdisjoint(s2))
```

```python
# frozenset
"""
A frozen set is in-effect an immutable set. You cannot change its values.
Also, a set can’t be used a key for a dictionary, but a frozenset can.
"""
```

```python
vowels = ('a', 'e', 'i', 'o', 'u')

fSet = frozenset(vowels)

print('The frozen set is:', fSet)

print('The empty frozen set is:', frozenset())
```

```python
# When you use a dictionary as an iterable for a frozen set, it only takes keys of the dictionary to create the set.
```

```python
# random dictionary
person = {"name": "John", "age": 23, "gender": "male"}

fSet = frozenset(person)
print('The frozen set is:', fSet)
```

```python

person = {"name": "Jack", "age": 21, "Country": "India"}

frozenDict = frozenset(person)

# remove comment and observe 
frozenDict[1] = "Work"
print('The frozenset object:', frozenDict)
```

### Password Strength Check using Sets

```python
import string

string.punctuation
```

```python
password = "Hello@123"

import string

uppercase = set("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
lowercase = set("abcdefghijklmnopqrstuvwxyz")
digits = set("0123456789")
specials = set(string.punctuation)
```

```python
pwd_chars = set(password)

has_upper = not uppercase.isdisjoint(pwd_chars)
has_lower = not lowercase.isdisjoint(pwd_chars)
has_digit = not digits.isdisjoint(pwd_chars)
has_special = not specials.isdisjoint(pwd_chars)
```

```python
if has_upper and has_lower and has_digit and has_special:
    
    print(" Password is strong!")
else:
    print(" Password is weak.")
```

### Find Pairs with Equal Sum in Two Lists

```python

def find_pairs_with_sum(a, b, target):
    
    return [(x, y) for x in a for y in b if x + y == target]

print(find_pairs_with_sum([1, 2, 3, 4], [5, 6, 7], 8))
```

#### Remove Stopwords Using Sets

```python
text = "this is a simple text with some common stopwords like and the or in"

stopwords = {"is", "a", "with", "some", "like", "and", "the", "or", "in"}

words = text.split()

print(words)
```

```python
filtered_words = set(words) - stopwords
print(filtered_words)
```
