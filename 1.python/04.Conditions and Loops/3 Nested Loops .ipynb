### Pattern 1

```python
for i in range(1, 6):
  
    for j in range(1, 6):
       
        print(i * j, end=' ')
    
    print()
```

### Pattern 2

```python
for i in range(1, 6):
    
    for j in range(1, 6):
        
        print(f"{i * j}".ljust(2), end=" ")
        
    print()
```

### Pattern 3

```python
rows = 5

for i in range(1, rows + 1):
    
    for j in range(1, i + 1):
        
        print("*", end=" ")
        
    print()
```

### Pattern 4

```python
for i in range(3):
    
    for j in range(3):
        
        if j == i:
            break
            
        print(i, j)
```

### Pattern 5

```python
for i in range(3):
    
    for j in range(3):
        
        if j == i:
            
            continue
            
        print(i, j)
```

### Pattern 6

```python
for i in range(1,6):
    print(str(i) * i)
```
