### Example Slicing

```python
s = input('Enter a string of at least 6 characters: ')

# (a) Total number of characters
print('(a) Length:', len(s))

# (b) Second character (index 1, since indexing starts at 0)
print('(b) 2nd char:', s[1])

# (c) Last character (index -1 refers to the last element)
print('(c) Last char', s[-1])

# (d) First 5 characters (from start up to index 5, not including index 5)
print('(d) First 5 chars:', s[:5])

# (e) Last 2 characters
print('(e) Last 2 chars:', s[-2:])

# (f) Reversed string (step = -1)
print('(f) Backwards:', s[::-1])

# (g) All except last character
print('(g) All chars except last:', s[:-1])

# (h) All except first and last characters
print('(h) All chars except first and last:', s[1:-1])
```

Write a function to add 'ing' at the end of a given string (length should be at least 3).
If the given string already ends with 'ing' then add 'ly' instead. 
If the string length of the given string is less than or equal to 2 chars, leave it unchanged.

```python
def add_string(str1):

    length = len(str1)
    if length < 3:
        return str1
        
    elif str1[-3:] == 'ing':
        return str1 + 'ly'
        
    else:
        return str1 + 'ing'
```

```python
print(add_string('ab'))
```

```python
print(add_string('abc'))
```

```python
print(add_string('string'))
```

Python function which accepts a list of strings containing details of deposits and withdrawals made in a bank account and returns the net amount in the account. 

Suppose the following input is supplied to the function ["D:300","D:300","W:200","D:100"] where D means deposit and W means withdrawal, then the net amount in the account is 500.

```python
def str_check(s):
    d=w=0
    
    for s1 in s:
        s2=s1.split(":")
        print(s2)
        
        if s2[0]=="D":
            dep= int(s2[1])
            d = d + dep
        else:
            wit = int(s2[1])
            w = w + wit
            
    return(d-w)
```

```python
print(str_check(["D:300","D:300","W:200","D:100"]))
```

```python
### alternative
```

```python
def str_check(s):
    
    d = w = 0
    for item in s:
        
        tx, amount = item.split(":")
        amount = int(amount)
        if tx == "D":
            d += amount
        else:
            w += amount
            
    return d - w
```

```python
print(str_check(["D:300","D:300","W:200","D:100"]))
```

#### Printing NUMBER OF TIMES substr repeated 

```python
def count_substring(string, sub_string):
   
    count = 0
    i = string.find(sub_string)

    while i != -1:
        count += 1
        i = string.find(sub_string, i+1)
    return count

string = 'ABCDCDC'
sub_string = 'CDC'

count = count_substring(string, sub_string)
print(count)
```

### Print counts with positions 

```python
def count_substring(string, sub_string):
    count = 0
    positions = []
    
    i = string.find(sub_string)
    while i != -1:
        count += 1
        positions.append(i)
        i = string.find(sub_string, i+1)  # allows overlapping matches
        
    return count, positions

string = 'ABCDCDC'
sub_string = 'CDC'

count, positions = count_substring(string, sub_string)
print("Count:", count)
print("Positions:", positions)
```

#### Counting no of occurances of a substring 

```python
s = "ABCDCDC"
sub = 'CDC'
 
count = 0
n = len(s)
m = len(sub)

for i in range(n - m + 1):
    if s[i:i+m] == sub:
        count += 1
 
print(count)
```

## checking for isalpha()  isdigit()

python prg, which accepts a sentence and finds the number of letters and digits in the sentence.

```python
s = 'year 2025'
d=l=0

for c in s:
    if c.isdigit():
        d=d+1
    elif c.isalpha():
        l=l+1
    else:
       continue
        
print("Letters", l)
print("Digits", d)
```

```python
s = 'year 2025 @home!'

letters = digits = spaces = specials = 0

for c in s:
    if c.isalpha():
        letters += 1
    elif c.isdigit():
        digits += 1
    elif c.isspace():
        spaces += 1
    else:
        specials += 1

print("Letters:", letters)
print("Digits:", digits)
print("Spaces:", spaces)
print("Special characters:", specials)
```

### Anagram

An anagram is a word or phrase formed by rearranging the letters of a different word 

```python
def is_anagram(s1,s2):
    if(sorted(s1)==sorted(s2)):
         return True
    else:
         return False

s1="SILENT"
s2="LISTEN"

is_anagram(s1,s2)
```
