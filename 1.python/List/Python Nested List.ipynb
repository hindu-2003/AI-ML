###  Python Nested Lists

```python
lst = ['a', 'b', ['c', 'd', ['e', 'f']], 'g']

print(lst[0])
print(lst[2])
print(lst[2][0])
print(lst[2][2])
print(lst[2][2][0])
print(lst[2][2][1])
```

```python
lst = [100,'python',[200,300],50.8,'a']  
lst[2][0] = 200.50
print (lst)
```

```python
lst2 = [100,'python',50.8,'a']  
lst3 = [200,2500.55]
lst2.append(lst3)
print (lst2)
```

```python
lst2 = [100,'python',50.8,'a']  
lst3 = [200,2500.55]
lst2.extend(lst3)
print (lst2)
```

```python
lst = [100,'python',[200,300],50.8,'a']  
print(lst)

del lst[2]
print(lst)
```

```python
lst = [100,'python',[200,300],50.8,'a']  
print(lst)

del lst[0:2]
print(lst)
```

```python
lst = [100,'python',[200,300],50.8,'a']  
del lst
print(lst)
```

```python
lst = [100,'python',[200,300],50.8,'a']  
del lst[2][0]
print(lst)
```

```python
lst = [100,'python',[200,300],50.8,'a']  
print(lst[0:2])
print(lst[2:3])
print(lst[:])
```

```python
lst = [100,'python',[200,300],50.8,'a']  
print(lst[-3:-5])
```

```python
print(lst[-5:-3])
```

```python
print(lst[-5:-1])
```

```python
lst = [100,'python',[200,300],50.8,'a']  

print(lst[-1:])
print(lst[:-1])
```

```python
lst1 = [1, 2, 3]
print(lst1 + [4, 5, 6])
```

```python
print(lst1)
```

```python
lst1 = [1, 2, 3]
l2 = lst1 + [4, 5, 6]
print(l2)
```

```python
lst = ['a', ['b', 'c'], 'd']
lst[1].append('python')
print(lst)
```

```python
lst = ['a', ['b', 'c'], 'd']
lst[1].insert(0,'python')
print(lst)
```

```python
lst = ['a', ['b', 'c'], 'd']
lst[1].extend("python")
print(lst)
```

```python
lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print(lst[0])
```

```python
print(lst[1])
```

```python
print(lst[1][0])
```

```python
print(lst[1][0][0])
```
