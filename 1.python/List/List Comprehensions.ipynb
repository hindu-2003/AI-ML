### List Comprehension

```python
lc1 = [x**2 for x in range(0,11)] 
print(lc1)                    
```

```python
lc2 = [2**i for i in range(5)]
print(lc2)                       
```

```python
lc3 = [x for x in lc1 if x % 2 == 0]
print(lc3)    
```

```python
lc4 = [x for x in lc1 if x % 2 != 0]
print(lc4)    
```

#### Even numbers in range of 0 to 10 and multiply

```python
even_squares = []
for num in range(0,11):
    if num%2 == 0:
         even_squares.append(num * num)
print(even_squares)
```

```python
even_squares = [num * num for num in range(0,11) if num%2 == 0]
even_squares
```

### Looping with individual chars

```python
chars = []
for letter in 'Comprehensions':
     chars.append(letter)
print(chars)
```

```python
chars = [letter for letter in 'Comprehensions']
print( chars)
```

### List comprehension with List of Strings

```python
# converting upper case chars
com = ["allow","sequences","to","be","built","from","other","sequences"]
lc4= [c.upper() for c in com]
lc4
```

### List comprehension with tuples

```python
# Convert height to feet  1 cm = 0.0328 feet

h_cms = [('akshay',183),('rahul',171),('sourav',179),('virat',190),('rohit',165)]

h_feet = [(height[0],round(height[1]*0.0328,2)) for height in h_cms]

print(h_feet)
```

### Nested Lists

```python
l1 = []
for x in [20, 40, 60]:
    for y in [2, 4, 6]:
        l1.append(x * y)
print(l1)
```

#### converting nested list to list comprehension

```python
l1 = [20,40,60]
l2 = [2,4,6]
l1 = [x * y for x in l1 for y in l2]
print(l1)
```

### Finding common elements in list

```python
lst_a = [1, 2, 3, 4, 6]
lst_b = [2, 3, 4, 5, 7, 2]

c_lst = [a for a in lst_a for b in lst_b if a == b]

print(list(c_lst))
```
