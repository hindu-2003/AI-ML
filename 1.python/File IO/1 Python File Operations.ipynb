#### Extract Txt  files

```python
import os
fn = []

for x in os.listdir():
    if x.endswith(".txt"):
        fn.append(x)
        
print(fn)
```

```python
data = []

for var in fn:
    with open(var, 'r', encoding='utf-8') as f:
        str1 = f.read()
        data.append(str1)
        
print(data)
```

```python
with open("output.txt", "w", encoding="utf-8") as f:
    for i in range(len(data)):
        f.write(data[i] + "\n\n")

print("All text written to output.txt")
```

### Read all text files in a directory

```python
import os

# Directory with your text files
text_dir = "./file/"

all_texts = {}

for filename in os.listdir(text_dir):
    if filename.lower().endswith(".txt"):
        filepath = os.path.join(text_dir, filename)
        print(filepath)
    
        with open(filepath, "r", encoding="utf-8") as f:
            content = f.read()

        all_texts[filename] = content


print(all_texts)

for fname, content in all_texts.items():
    print(f"\n {fname}:\n{content[:200]}")
```

### merge contents in to single file

```python
import os


text_dir = "./file/"
output_file = "merged_output.txt"

with open(output_file, "w", encoding="utf-8") as outfile:
    for filename in os.listdir(text_dir):
        if filename.lower().endswith(".txt"):
            filepath = os.path.join(text_dir, filename)

            with open(filepath, "r", encoding="utf-8") as infile:
                content = infile.read()

            outfile.write(f"\n--- {filename} ---\n")
            outfile.write(content + "\n")

print(f" All text files merged into {output_file}")
```
