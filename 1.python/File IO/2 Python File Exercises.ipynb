### The file expenses.txt has a bunch of expenses, each an integer, one on each line. 

Write a program that reads the file and prints out how many expenses are over $2000. 

Print all of them on the same line

```python
lines = [line.strip() for line in open('./contents/expenses.txt')]

print(lines)
```

```python
exp = [int(item) for item in lines if int(item) > 2000]

print(exp)
```

### usimg map

```python

with open('./contents/expenses.txt', "r") as f:
    lines = [line.strip() for line in f]
print(lines)
```

```python

numbers = list(map(int, lines))
print(numbers)
```

```python

exp = [num for num in numbers if num > 2000]
print(exp)
```

### Use wordlist.txt to do the following.

#### All words that contain the string "quot".

```python
words = [line.strip() for line in open('./contents/wordlist.txt')]
print(words[:50])
```

```python
for word in words:
    if 'quot' in word:
        print(word, end=' ')
print()
```

### All words of at least six letters that start and end with the same three letters.

```python

for word in words:
    if len(word) >= 6 and word[:3] == word[-3:]:
        print(word, end=' ')
print()        
```

### The longest word that starts and ends with the same letter.

```python

longest = 'a'

for word in words:
    if word[0] == word[-1] and len(word) > len(longest):
        longest = word

print(longest)
```

### Write a program that reads through wordlist.txt and produces a new file new_wordlist.txt 

that contains all the words from wordlist.txt that are 5 or less letters long.

```python
output_file = open('new_wordlist.txt', 'w')

lines = [line.strip() for line in open('./contents/wordlist.txt')]
print(lines[:10])
```

```python
for line in lines:
    if len(line) <= 5:
        print(line, file=output_file)
output_file.close()
print("end")
```

#### Each line of the file population.txt contains a country name and its population, separated by a tab.

Write a program that uses the file to print out all the countries whose names start with G that have at least 500,000 people.

```python
lines = [line.strip() for line in open('./contents/population.txt')]
print(lines)
```

```python
for line in lines:
    L = line.split('\t')
    country = L[0]
    population = int(L[1])
    
    if country[0] == 'I' and population >= 5000000:
        print(country, population)
```
